## **Part 2. Cleaning data**

```python
# Dropping duplicates

food.drop_duplicates(keep=False, inplace=True)
```

```python
# Checking duplicates:

food.info()

# We succesfully deletes 382 duplicates 
# No null in our date
```

```python
# Creating new Column 
food['Total_Children']=food[['Kidhome','Teenhome']].sum(axis=1)
```

```python
# Transformed one-hot encoded marital status variables into a single ordinal representation by assigning weighted numeric values to each category.

food['marital_Divorced']=food['marital_Divorced'].replace({1:5,0:0})
food['marital_Married']=food['marital_Married'].replace({1:4,0:0})
food['marital_Single']=food['marital_Single'].replace({1:3,0:0})
food['marital_Together']=food['marital_Together'].replace({1:2,0:0})
food['marital_Widow']=food['marital_Widow'].replace({1:1,0:0})
```

```python
# Combining the marital status columns into one column by adding them together to create a single “Marital Status” variable.

food['Marital Status']=food[['marital_Divorced','marital_Married','marital_Single', 'marital_Together', 'marital_Widow']].sum(axis=1)
```

```python
# Converting the numeric marital status values into text labels to make the data easier to understand.
food['Marital_Status_str']=food['Marital Status'].map({5: 'Divorced', 4:'Married', 3:'Single', 2:'Together', 1:'Widow'})
```

```python
#I am assigning numeric values to each education category to create an ordered scale based on education level.

food['education_2n Cycle']=food['education_2n Cycle'].replace({1:1,0:0})
food['education_Basic']=food['education_Basic'].replace({1:2,0:0})
food['education_Graduation']=food['education_Graduation'].replace({1:3,0:0})
food['education_Master']=food['education_Master'].replace({1:4,0:0})
food['education_PhD']=food['education_PhD'].replace({1:5,0:0})

```

```python
#I am combining the education columns into one column by adding them together to create a single “Education Status” variable.

food['Education_Status']=food[['education_2n Cycle','education_Basic','education_Graduation', 'education_Master', 'education_PhD']].sum(axis=1)
```

```python
# Adding the campaign columns together to create a single variable that shows how many campaigns each customer accepted.
food['Accepted_Campaigns']=food[['AcceptedCmp3','AcceptedCmp4','AcceptedCmp5', 'AcceptedCmp2', 'AcceptedCmp1']].sum(axis=1)
```

```python
food['Accepted_Campaigns']=(food['Accepted_Campaigns']!=0).astype(int)
```

## **3.1 Age Analysis**

The heatmap shows that people who spent more money and accepted campaigns before were more likely to accept campaigns again. It also shows that age, marital status, and education had little connection with campaign acceptance. Overall, customer behavior was much more useful than personal background for understanding who accepted campaigns.

```python
# This code assigns each customer to an age range and creates a new column with those groups
age_groups= [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group (Age):
    for age_range in age_groups:
        if age_range[0]<= Age <= age_range[1]:
            return f"{age_range[0]}-{age_range[1]}"
    return ("Unknown")

food["Age_Group"]= food["Age"].apply(assign_age_group)
```

```python
# This code plots the average campaign acceptance for each age group in a specific order

age_order=['23-30','31-40','41-50','51-60','61-70','71-85']

sns.pointplot(data= food, x='Age_Group', y= 'Accepted_Campaigns', order= age_order)
plt.title('Average Campaign Acceptance Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Accepted Campaigns')
plt.show()
```

```python
# This code calculates the percentage of customers in each age group and sorts them in order
counts=food['Age_Group'].value_counts()

percentage= counts/food.shape[0]

percent_food=percentage.reset_index()

percent_food.columns=['age_group', 'percentage']

percent_food= percent_food.sort_values(by='age_group', ascending= True)

percent_food
```

The chart shows that younger (23–30) and older (71–85) customers have higher campaign acceptance rates, but these groups represent a very small percentage of the total customers, so the results may not be very reliable. In contrast, most customers are in the middle age groups (especially 41–60), where acceptance rates are lower, making these results more representative of the overall customer base.

```python
# This code creates a bar chart showing the percentage of customers in each age group

sns.barplot(x='age_group', y='percentage', data= percent_food)
plt.title('Percentage of Accepted Campaigns per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.show()
```

The chart shows that younger (23–30) and older (71–85) customers have higher campaign acceptance rates, but these groups are very small, so the results may not be very reliable. Most customers are in the middle age groups (especially 31–70), where acceptance rates are lower but more representative of the overall customer base.


```python
# This code calculates total spending per age group and visualizes it with a bar chart

grouped_food=food.groupby('Age_Group')['MntTotal'].sum().reset_index()

sns.barplot(x='Age_Group', y='MntTotal', data= grouped_food)
plt.title('Total Customer Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Spending ($)')

plt.show()
```

The chart shows that customers aged 41–50 spend the most money, followed by those in the 51–60 and 61–70 groups, while younger (23–30) and older (71–85) customers spend significantly less. This suggests that middle-aged customers are the most valuable in terms of total spending.

```python
# This code filters customers who accepted campaigns and plots total spending per age group

acct_camp= food[food['Accepted_Campaigns']!=0]

grouped_food=acct_camp.groupby('Age_Group')['MntTotal'].sum().reset_index()

sns.barplot(x='Age_Group', y='MntTotal', data= grouped_food)
plt.title('Total Spending by Age Group (Customers Who Accepted Campaigns)')
plt.xlabel('Age Group')
plt.ylabel('Total Spending ($)')
```

Compared with the previous chart, customers aged 41–50 are the most valuable group, with both the highest campaign acceptance and total spending. Ages 51–70 also spend a lot but respond less to campaigns. Younger (23–30) and older (71–85) groups show low engagement and low spending, so they are less important for revenue.ally 41–60) spend the most, even though they accept fewer campaigns, meaning they are more valuable customers overall despite lower engagement.

23-30, 71 and up accept are higher rates.

## **3.2 Channel Analysis**

```python
# This code creates a table with total purchases by channel and formats it for easier read

sum_food=pd.DataFrame(food[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(), columns=['Sums'])

sum_food= sum_food.reset_index()

sum_food.rename(columns={'index':'Type_of_Purchase'}, inplace= True)

```

```python
# This code creates a bar chart comparing total purchases across different channels
sns.barplot(x='Type_of_Purchase', y='Sums',hue='Type_of_Purchase', data=sum_food, palette=['lightblue', 'lightgreen', 'red'])

plt.title('Total Purchases by Channel')
plt.xlabel('Purchase Channel')
plt.ylabel('Number of Purchases')

plt.show()
```

```python
# This code analyzes customers who accepted campaigns and compares their purchases across different channels 
acct_camp= food[food['Accepted_Campaigns']!=0]

sum_food=pd.DataFrame(acct_camp[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(), columns=['Sums'])

sum_food= sum_food.reset_index()

sum_food.rename(columns={'index':'Type_of_Purchase'}, inplace= True)

sns.barplot(x='Type_of_Purchase', y='Sums', hue='Type_of_Purchase',
    data=sum_food,
    palette=['lightblue', 'lightgreen', 'red'],
    legend=False
)

plt.title('Purchases by Channel (Campaign Responders)')
plt.xlabel('Purchase Channel')
plt.ylabel('Number of Purchases')

plt.show()
```

```python
sum_food
```

Compared to all customers, the group that accepted campaigns still shows the same pattern: store purchases are the highest, followed by web purchases, and then catalog purchases. This means that even customers who accept campaigns prefer buying in-store, so physical stores remain the most important channel overall.

```python
# This code shows the relationship between total spending and web purchases using a density plot with a regression line

x= sns.jointplot(data=food, x='MntTotal', y='NumWebPurchases', kind='kde')
x.plot_joint(sns.regplot, color= 'red')
x.fig.suptitle('Total Spending vs. Web Purchases', y=1.05)
x.set_axis_labels('Total Spending ($)', 'Number of Web Purchases')
```

The chart shows a positive relationship between total spending and number of web purchases, meaning customers who buy more often online tend to spend more overall. Most customers are concentrated in lower spending and fewer purchases, with fewer high-spending customers making many web purchases.

```python
# This code shows the relationship between total spending and catalog purchases using a density plot with a regression line

x= sns.jointplot(data=food, x='MntTotal', y='NumCatalogPurchases', kind='kde')
x.plot_joint(sns.regplot, color= 'red')
x.fig.suptitle('Total Spending vs. Catalog Purchases', y=1.05)
x.set_axis_labels('Total Spending ($)', 'Number of Catalog Purchases')
```

The chart shows a positive relationship between total spending and catalog purchases, meaning customers who make more catalog purchases tend to spend more overall. Most customers are concentrated at lower levels, but higher spending is generally linked to more catalog purchases.

```python
# This code shows the relationship between total spending and store purchases using a density plot with a regression line
x= sns.jointplot(data=food, x='MntTotal', y='NumStorePurchases', kind='kde')
x.plot_joint(sns.regplot, color= 'red')
x.fig.suptitle('Total Spending vs. Store Purchases', y=1.05)
x.set_axis_labels('Total Spending ($)', 'Number of Store Purchases')
```

The chart shows a strong positive relationship between total spending and store purchases, meaning customers who buy more in-store tend to spend significantly more overall. Compared to web and catalog, store purchases appear to have the strongest connection with total spending.

**Highlight** 

We have two directions: either focus on increasing engagement with catalog customers, who show strong spending potential, or prioritize store and web channels, which have higher customer traffic and overall volume.

```python
# Calculates campaign acceptance % by marital status and plots it
total= food['Marital_Status_str'].value_counts()

accepted=food[food['Accepted_Campaigns']==1]['Marital_Status_str'].value_counts()

pect_marital= accepted/total * 100

pect_food=pect_marital.reset_index()

pect_food.columns=['Marital_Status','Percentage']

sns.barplot(x= 'Marital_Status', y= 'Percentage', hue='Marital_Status',
    data=pect_food,
    palette=['red', 'lightblue', 'pink', 'orange', 'green'])
plt.title('Campaign Acceptance Rate by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Acceptance Rate (%)')

plt.show()
```

The chart shows that widowed customers have the highest campaign acceptance rate, followed by single and married customers, while divorced customers have the lowest. However, since widowed customers represent a small group, the results may be less reliable, and focusing on larger groups like married and single customers may be more effective.

# **Part 4. Overall Findings**

#### **4.1. Age**
Customers between ages 30 and 70 generated the highest total spending, making them the most valuable segment from a revenue perspective. However, younger customers (23–30) and older customers (71–85) showed higher campaign acceptance rates. Because those two groups are much smaller in size, those results should be interpreted with caution.

#### **4.2. Purchase Channel**
Catalog customers appeared more likely to respond to campaigns, which suggests stronger campaign engagement in that channel. At the same time, in-store purchases generated the highest overall purchase volume, followed by web purchases. This suggests the business should not rely on only one channel. A balanced strategy across catalog, store, and web is more appropriate.

##### **Recommended channel mix**
- 40% catalog
- 30% store
- 30% web

#### **4.3. Children in Household**
Customers with no children, or fewer children, tended to spend more and were also more likely to accept campaigns. This suggests household size may influence both spending behavior and marketing responsiveness.

#### **4.4. Education**
Education showed little to no meaningful impact on spending or campaign acceptance. Based on this analysis, education level should not be a primary targeting variable.

#### **4.5. Marital Status**
Marital status did not appear to strongly influence campaign acceptance overall. However, married, single, and together customers contributed more total spending than divorced and widowed customers. This makes them more important from a revenue perspective.



# **Part 5. Business Recommendations**

#### **5.1. Revenue Growth Strategy**
The strongest revenue opportunity appears to be middle-aged, higher-spending customers, especially those with no children or fewer children. These customers should be targeted across multiple channels, with greater attention given to the recommended channel mix above.

#### **5.2. Customer Acquisition Strategy**
To increase campaign response and attract more engaged customers, the company should also target younger (23–30) and older (71–85) customers, since these groups showed higher campaign acceptance rates. However, because these groups are smaller, they should be treated as testing or growth segments rather than the core revenue segment.

#### **5.3. Final Conclusion**
The best business approach is to separate customers into two goals: revenue generation and campaign responsiveness. Middle-aged customers drive spending, while younger and older customers appear more likely to respond to campaigns. A strong strategy would target high-spending middle-aged customers for revenue, while using campaigns to test and grow the more responsive younger and older segments.
